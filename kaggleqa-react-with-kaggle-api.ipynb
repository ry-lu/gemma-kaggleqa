## Introduction
Inspired from the [ReAct Paper](https://arxiv.org/abs/2210.03629), this notebook aims to show how to leverage existing APIs to assist LLMs in answering common questions regarding Kaggle and its platform.

**ReAct** prompting is when the model employs reasoning and actions to complete a task. In this notebook, our Gemma model will utilize reasoning and leverage the Kaggle API along with Kaggle documents to answer Kaggle questions. This approach enables the model not only to search for information in the documents to respond to questions, thus grounding its responses, but also to access the current Kaggle datasets/kernels with the API.

The user will be able to see the model's thought process and the sources it used to lead to the response.

When we call the Kaggle API for lists of datasets/kernels/competitions, it returns objects. We need to access these objects and return relevant info.

````python
def get_dataset_info(dataset):
    """Retrieve relevant information from a dataset object.

    Args:
        dataset: A dataset object obtained from the Kaggle API.

    Returns:
        A formatted string containing the following dataset information:
            - Dataset Title
            - Subtitle
            - License
            - Kernel Count
            - URL
    """
    dataset_info = """ 
    Dataset Title: {title}
    Subtitle: {subtitle}
    License: {license}
    Kernel Count: {kernel_count}
    URL: {url}
    """
    return dataset_info.format(
        title=dataset.titleNullable,
        subtitle=dataset.subtitleNullable,
        url=dataset.urlNullable,
        license = dataset.licenseNameNullable,
        kernel_count = dataset.kernelCount
    )

def get_competition_info(competition):
    """Retrieve relevant information from a competition object.

    Args:
        competition: A competition object obtained from the Kaggle API.

    Returns:
        A formatted string containing the following competition information:
            - Competition Title
            - Description
            - Category
            - Evaluation Metric
            - Team Count
            - Reward
            - URL
    """
    
    competition_info = """ 
    Competition Title: {title}
    Description: {description}
    Category: {category}
    Evaluation Metric: {evaluation_metric}
    Team Count: {team_count} 
    Reward: {reward}
    URL: {url}
    """
    return competition_info.format(
        title=competition.titleNullable,
        description=competition.descriptionNullable,
        category = competition.categoryNullable,
        evaluation_metric=competition.evaluationMetricNullable,
        team_count = competition.teamCount,
        url=competition.url,
        reward=competition.rewardNullable
    )

def get_kernel_info(kernel):
    """Retrieve relevant information from a kernel object.

    Args:
        kernel: A kernel object obtained from the Kaggle API.

    Returns:
        A formatted string containing the following kernel information:
            - Kernel Title
            - Author
            - Votes
            - Ref
    """
    
    kernel_info = """ 
    Kernel Title: {title}
    Author: {author}
    Votes: {totalVotes}
    Ref: {ref}
    """
    return kernel_info.format(
        title=kernel.title,
        author=kernel.author,
        totalVotes=kernel.totalVotes,
        ref=kernel.ref
    )
````

## Kaggle Env


For the Kaggle Env, we have the following actions to assist in answering common Kaggle questions:

- **search_datasets[term]** which searches based on the term the public Kaggle datasets and returns a list of datasets.
- **search_kernels[term]** which searches based on the term the public Kaggle kernels and returns a list of kernels.
- **search_docs[term]** which returns sections that have the exact term in the Kaggle Documents.
- **list_comps[]** which lists all ongoing Kaggle competitions and returns a list of competition info.
- **finish[answer]** which returns the answer and finishes the task.

````python
valid_actions = ["list_comps[], search_datasets[]","list_comps[]","search_kernels[]","search_docs[]"]
class KaggleEnv():
    """Kaggle Environment class for interacting with Kaggle API.

    Interface to interact with the Kaggle API to perform various actions
    such as searching datasets, listing competitions, searching kernels,
    and searching documents.

    Attributes:
        obs (str): Current observation or state of the environment.
        answer (str): Answer obtained from the environment (if any).

    Methods:
        reset(): Reset the environment to its initial state.
        search_datasets(term): Search for datasets based on a given term using Kaggle API.
        list_competitions(): List ongoing competitions using Kaggle API.
        search_kernels(term): Search for kernels based on a given term using Kaggle API.
        search_docs(term): Locate and return document sections with a given term in the middle.
        step(action): Take a step in the environment based on the provided action.

    """
    def __init__(self):
        super().__init__()
        self.obs = None
        self.answer = None
        
    def reset(self):
        self.obs = "Interact with Kaggle API using search_datasets[], list_comps[], search_kernels[], and finish[].\n"
        self.answer = None
    
    @staticmethod
    def search_datasets(term):
        """Searchs for datasets based on term  using Kaggle API."""
        datasets = kaggle.api.dataset_list(search=term)[:10]
        if datasets:
            datasets_meta = [get_dataset_info(dataset) for dataset in datasets]
            return "\n ---- \n".join(datasets_meta)
        return "No datasets found. Maybe use a more broad search term. Search terms should be no longer than 2 words."
    
    @staticmethod
    def list_competitons():
        """Lists ongoing competitions"""
        competitions = kaggle.api.competitions_list()
        if competitions:
            competitions_meta = [get_competition_info(competiton) for competiton in competitions]
            return "\n ---- \n".join(competitions_meta)
        return "No ongoing competitions found."
    
    @staticmethod
    def search_kernels(term):
        """Searchs for kernels based on term using Kaggle API."""
        kernels = kaggle.api.kernels_list(search=term)[:10]
        if kernels:
            kernels_meta = [get_kernel_info(kernel) for kernel in kernels]
            return "\n ---- \n".join(kernels_meta)
        return "No kernels found. Maybe use a more broad search term. Search terms should be no longer than 2 words."
    
    @staticmethod
    def search_docs(term):
        """Locates term and returns sections with term in the middle."""
        sections = []
        i = 0
        while i < len(kaggle_docs):
            if term in kaggle_docs[max(0, i - len(term)):min(len(kaggle_docs), i + len(term))]:
                # Extract section
                sections.append(
                    kaggle_docs[max(0, i - 500):min(len(kaggle_docs), i + 500)]
                )
                # Skip over section in docs
                i += 500
            else:
                i += 1

        if sections:
                return "\n ---- \n".join(sections[:5])

        return "No documents found. Maybe use a more broad search term. Search terms should be no longer than 2 words."
       
        
    def step(self, action):
        """Take a step in the environment based on the provided action.

        Args:
            action (str): Action to be performed in the environment.

        Returns:
            tuple: A tuple containing:
                - obs (str): Current observation or state of the environment.
                - done (bool): True if the question is answered, otherwise False.
                - invalid (bool): True if the action is invalid, otherwise False.

        """
        done = False # Is question answered
        invalid = False # Is invalid action
        action = action.strip()
        
        if self.answer is not None:
            done = True
            return self.obs, done
        
        # Parse out actions
        if action.startswith("search_datasets[") and action.endswith("]"):
            term = action[len("search["):-1]
            self.obs = self.search_datasets(term)
    
        elif action.startswith("list_comps[") and action.endswith("]"):
            term = action[len("search["):-1]
            self.obs = self.list_competitons()
            
        elif action.startswith("search_kernels[") and action.endswith("]"):
            term = action[len("search["):-1]
            self.obs = self.search_kernels(term)
            
        elif action.startswith("search_docs[") and action.endswith("]"):
            term = action[len("search_docs["):-1]
            self.obs = self.search_docs(term)
            
        elif action.startswith("finish["):
            answer = action[len("finish["):-1]
            self.answer = answer
            done = True
            self.obs = f"Finished."
            
        else:
            self.obs = "Invalid action: {action}. Close your brackets []. Your valid actions are {valid_actions}".format(action=action, valid_actions=valid_actions)
            invalid = True
        return self.obs, done, invalid
````

# Designing the LLM Agent

Like in the ReAct paper, our LLM Agent should take actions to help respond to the common Kaggle question. This will involve prompting and parsing its output.

````python
# For Gemma
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, StoppingCriteria
import torch
from peft import AutoPeftModelForCausalLM

# For Gemini
import google.generativeai as genai

device = "cuda" if torch.cuda.is_available() else "cpu"
````

````python
class MyStoppingCriteria(StoppingCriteria):
    """Stops generation when encountering target_sequence."""
    def __init__(self, target_sequence, prompt):
        self.target_sequence = target_sequence
        self.prompt=prompt

    def __call__(self, input_ids, scores, **kwargs):
        generated_text = tokenizer.decode(input_ids[0])
        generated_text = generated_text.replace(self.prompt,'')
        if self.target_sequence in generated_text:
            return True 

        return False  

    def __len__(self):
        return 1

    def __iter__(self):
        yield self

class LLMAgent:
    """Agent for interacting with Kaggle Environment using LLM (Large Language Model).

    Attributes:
        env (KaggleEnv): The Kaggle Environment instance.
        model: The LLM model for generation.
        instruction (str): Instruction for resolving a Kaggle Assistant task with interleaving Thought, Action, Observation steps.
        is_gemma (bool): Flag indicating whether to use Gemma for generation.
        examples (str): Examples to be added to the instruction.
    
    Methods:
        use_llm_gemma(prompt, stop=["\n"]): Generate output using Gemma.
        use_gemini(prompt, stop=["\n"]): Generate output using Gemini.
        generate_answer(question, logging=True): Generate answer to the instruction.

    """
    def __init__(self,env, examples, is_gemma=True, add_examples = True):
        """Initialize LLMAgent."""
        super().__init__()
        self.env= env # Kaggle Env
        self.model = model
        self.instruction =  """Resolve a kaggle assistant task with interleaving Thought, Action, Observation steps. Thought can reason about the current situation, and Action can be three types: 
        (1) search_datasets[term] which searches based on the term the public Kaggle datasets and returns a list of datasets.
        (2) search_kernels[term] which searches based on the term the public Kaggle kernels and returns a list of kernels.
        (3) list_comps[] which lists all ongoing Kaggle competitions and returns a list of competition info.
        (4) search_docs[term] which returns all sentences that have the exact term in the Kaggle Documents (Notebooks.txt, Public API.txt, Organizations.txt, Competitions Setup.txt, Datasets.txt, Models.txt, TPU Usage.txt, GPU Usage.txt, Competitions.txt, kaggle-privacy-policy.txt, kaggle-terms-use.txt). Be broad with terms.
        (5) finish[answer] which returns the answer and finishes the task.
        """
        
        self.is_gemma=is_gemma
        if add_examples:
            self.instruction+=examples
    
    def use_llm_gemma(self, prompt,stop=["\n"]):
        """Uses Gemma to generate output."""
        inputs = tokenizer(prompt, return_tensors="pt").to(device)
        stopping_criteria = MyStoppingCriteria(stop[0], prompt)
        outputs = model.generate(**inputs, stopping_criteria=stopping_criteria, max_new_tokens=500)
        return tokenizer.decode(outputs[0], skip_special_tokens=True).replace(prompt,"")
        
    def use_gemini(self, prompt, stop=["\n"]):
        """Uses Gemini to generate output."""
        generation_config = {
            'stop_sequences':stop,
            'top_p':1,
            "max_output_tokens":500,
            'temperature':0,

        }

        response = model.generate_content(
            prompt,
            generation_config=generation_config,
        )
        try:
            return response.text
        except Exception as e:
            print(e)
            return ""
    
    def generate_answer(self, question,logging=True):
        """Generate an answer to the given instruction/question.

        Args:
            question (str): The instruction for which the answer is to be generated.
            logging (bool, optional): Flag indicating whether to log the steps in Markdown. Default is True.

        Returns:
            tuple: A tuple containing:
                - answer (str): The generated answer.
                - prompt_for_training (str): The trajectory used for finetuning.
                - markdown_text (str): The Markdown text of the generated trajectory.
        """

        # See what llm to use
        if self.is_gemma:
            use_llm = self.use_llm_gemma
        else:
            use_llm = self.use_gemini
            
        self.env.reset()
        markdown_text = "" # What to print out
        question = f'Instruction: {question}'
        prompt = self.instruction + question
        
        # Trajectory to use for finetuning
        prompt_for_training = self.instruction+question
        
        # Generate Thought Action Observation sequences
        for i in range(1,7):
            thought_action = use_llm(prompt + f"Thought {i}:", stop=["Observation"])
            try:
                thought_action = thought_action.replace("Observation","")
                thought, action = thought_action.strip().split(f"\nAction {i}: ")
            except:
                thought = thought_action.strip().split('\n')[0]
                action = use_llm(prompt + f"Thought {i}: {thought}\nAction {i}:", stop=[f"\n"]).strip()
            
            if logging:
                markdown_text += f"\n**Thought {i}**: {thought}\n"
                markdown_text +=f"\n**Action {i}**: {action}\n"
                
            obs, done, invalid = self.env.step(action)
            
            if logging:
                markdown_text += "```\n" + obs + "\n```"
                
            obs = obs.replace('\\n', '')
            step_str = f"Thought {i}: {thought}\nAction {i}: {action}\nObservation {i}: {obs}\n"
            prompt += step_str
            
            # We do not want to use invalid trajectories for training
            if not invalid:
                prompt_for_training+=step_str

            if done:
                break
            gc.collect()
            torch.cuda.empty_cache()
            
        if not done:
            obs, done, invalid = self.env.step("finish[]")
            
        return self.env.answer, prompt_for_training, markdown_text
````

# Conclusion
This notebook demonstrates how to use the Kaggle API alongside ReAct Prompting to build an LLM capable of retrieving information directly from Kaggle to answer questions. Gemma underwent fine-tuning using Gemini's outputs for knowledge distillation, resulting in interpretable results.

## Usefulness
ReAct prompting enhances model interpretability by grounding it in current, factual information. Retrieving information through existing Kaggle APIs simplifies the process, and the environment demonstrated here can be adapted for use with any LLM.

## Limitations
Despite the fine-tuning, Gemini remains significantly more robust than Gemma. Gemma may occasionally hallucinate or fail to provide a response. Due to hardware/time limitations, Gemma could only be fine-tuned on the 2b version with LoRA on under 1000 trajectories. To improve upon this work and create a more robust model, one could consider training a larger Gemma model on more trajectories. Another limitation is the context window; sometimes, search_docs may return documents larger than the context, causing Gemma to malfunction. A potential solution could involve using summarization tools to trim down the retrieved documents and shorten the context. Moreover, the model heavily relies on the observations obtained from the Kaggle API. Sometimes, the Kaggle API search returns dubious results or descriptions are misleading, interfering with the model's performance.

## Closing Remarks
This notebook only scratches the surface of what's possible with the ReAct framework and Kaggle API integration into Gemma. While it has potential, the current Gemma model still is not exactly reliable due to hardware/time constraints. There are numerous avenues for further improvement and exploration. Adding summarization, RAG for loading Kaggle documents on top of the Kaggle Environment are all possibilities for improvement.
